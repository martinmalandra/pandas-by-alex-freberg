En este caso podemos agrupar por continente.

```python
df.groupby('Continent').mean()
```

```python
df[df['Continent'].str.contains('Oceania')]
```

```python
df.groupby('Continent').mean().sort_values(by="2022 Population", ascending=False)
```

# Visualización para responder qué país creció más en población

```python
df2 = df.groupby('Continent').mean().sort_values(by="2022 Population", ascending=False)
df2
```

```python
df2.plot()
```

Para visualizar con mayor claridad, trasponemos los datos.

```python
df3 = df2.transpose() # los indices son las columnas
```

```python
df3.plot()
```

La gráfica tiene valores de columnas que no queremos en el eje X (Rank, Density y World Population Percentage). Con lo cual hacemos lo siguiente.

```python
df.columns # es para ganar tiempo copypasteando las columnas del DF
```

```python
df4 = df.groupby('Continent')[['1970 Population',
       '1980 Population', '1990 Population', '2000 Population',
       '2010 Population', '2015 Population', '2020 Population',
       '2022 Population']].mean().sort_values(by="2022 Population", ascending = False)
```

```python
df4 = df4.transpose()
```

```python
df4.plot()
```

Al estar al revés, tenemos que dar vuelta el orden de selección de las columnas en la línea 66.
