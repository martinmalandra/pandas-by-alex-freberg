# Indexado

```python
import pandas as pd
```

```python
df = pd.read_csv("world_population.csv")
df
```

```python
# Agregamos el pais como indice, forma 1

df = pd.read_csv("world_population.csv", index_col='Country')
df
```

```python
df.reset_index(inplace=True) # Reset de indice
df
```

```python
df.set_index('Country',inplace=True) # segunda forma de nombrar indice segun una columna
df
```

```python
# Busquedas por indices

# loc

df.loc['Argentina']
```

```python
# iloc

df.iloc[8]
```
