# Visualizaciones en Pandas

```python
print(plt.style.available)
plt.style.use('seaborn-v0_8-deep')
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Ice Cream Ratings.csv")
df = df.set_index('Date')
df
```

```python
df.plot() # por defecto nos da una grfica de lneas
```

```python
df.plot(kind='line') # por defecto nos da una grfica de lneas
```

```python
df.plot(kind='line', subplots=True) # separa en subgraficas por columnas
```

```python
df.plot(kind='line', title='Ice Cream Ratings', xlabel='Daily Rating', ylabel='Scores') # titulo y titulos de ejes
```

```python
df.plot(kind='bar') # grafico de barras
```

```python
df.plot(kind='bar', stacked=True) # grafico de barras apiladas
```

```python
df.plot.barh(stacked=True)
```

```python
df.plot.scatter(x='Texture Rating', y='Overall Rating')
```

```python
df.plot.scatter(x='Texture Rating', y='Overall Rating', s = 100)
```

```python
df.plot.scatter(x='Texture Rating', y='Overall Rating', s = 100, c='red')
```

```python
df.plot.hist()
```

```python
df.plot.hist(bins=20)
```

```python
df.boxplot()
```

```python
df.plot.area(figsize=(10,5))
```

```python
df.plot.pie(y='Flavor Rating', figsize=(10,6))
```
